# 4. Data Preparation

```python
user_pivot_table = pd.pivot_table(data = books_data,values = 'Book.Rating',index= 'User.ID',columns = 'Book.Title').fillna(0)
user_pivot_table.index = books_data['User.ID'].unique()
user_pivot_table
```

# 5. Model Building

## 5.1 Cosine Matric

```python
cosine_metrices = 1 - pairwise_distances(X = user_pivot_table, metric='cosine' ).round(3)
cosine_metrices_df = pd.DataFrame(cosine_metrices)
cosine_metrices_df.index = books_data['User.ID'].unique()
cosine_metrices_df.columns = books_data['User.ID'].unique()
cosine_metrices_df.head(25)
```

## From the cosine similarity recommendation model we can conclude there can be no recommendation done on the basis of similarity as the books are not similar in nature as far as we can see.

## For a better Recommendation system we can try different models as well.

## 5.2 Euclideon Distance

```python
euclideon_distance_user = pairwise_distances(X = user_pivot_table,metric='euclidean').round()
euclideon_distance_user_df = pd.DataFrame(data = euclideon_distance_user)
euclideon_distance_user_df.columns = books_data['User.ID'].unique()
euclideon_distance_user_df.index   = books_data['User.ID'].unique()
euclideon_distance_user_df
```

## In this model we look into the distance between two features or books, lesser the distance more correlated the items are and more chances for recommendation. For eg in the above model book number 276751 and 276729 have distance of 9 therefore they can be recommended to the user whereas book 276726 and 276736 have a distance of 16 therefore it cannot be recommended to the user.
